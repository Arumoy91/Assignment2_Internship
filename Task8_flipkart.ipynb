```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver=webdriver.Chrome("C:\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
driver.get(r'https://www.flipkart.com/search?q=sneakers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
```

```python
brand_titles=[]
price_mrp=[]
discount_titles=[]
```

```python
title_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
title_tags[0:100]
```

```python
for i in title_tags:
    
    title=i.text
    brand_titles.append(title)
brand_titles[0:100]
```

```python
price_mrp=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
price_mrp[0:100]
```

```python
discount_titles=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
discount_titles[0:100]
```
