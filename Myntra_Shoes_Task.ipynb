```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver=webdriver.Chrome("C:\chromedriver.exe")
```

```python
driver.get('https://www.myntra.com/shoes')
```

```python
driver.get('https://www.myntra.com/shoes?f=Color%3ABlack_36454f&plaEnabled=false&rf=Price%3A7649.0_15099.0_7649.0%20TO%2015099.0&sort=new')
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
```

```python
brand_titles=[]
```

```python
title_tags=driver.find_elements_by_xpath("//li[@class='product-base']")
title_tags[0:100]
```

```python
for i in title_tags:
    
    title=i.text
    brand_titles.append(title)
brand_titles[0:100]
```

```python
print(len(brand_titles))
```

```python
import pandas as pd
brand=pd.DataFrame({})
brand['brand_titles']=brand_titles[0:100]
```
